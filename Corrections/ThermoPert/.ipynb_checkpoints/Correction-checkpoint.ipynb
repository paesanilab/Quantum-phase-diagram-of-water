# Correction by thermodynamic pertubation

```python
import numpy as np
import glob
import matplotlib.pylab as plt
colors=['#377eb8','#4daf4a','#e41a1c','#984ea3','#ff7f00','#a65628','#f781bf','#999999']*4
SMALL_SIZE = 8
MEDIUM_SIZE = 10
BIGGER_SIZE = 12
plt.rc('font', size=MEDIUM_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
%config InlineBackend.figure_format = 'retina'
kb=8.617333262E-5
mev_to_kj=96.485307/1000.
import matplotlib
plt.rcParams['text.usetex'] = True
matplotlib.rcParams['mathtext.fontset'] = 'custom'
matplotlib.rcParams['mathtext.rm'] = 'Bitstream Vera Sans'
matplotlib.rcParams['mathtext.it'] = 'Bitstream Vera Sans:italic'
matplotlib.rcParams['mathtext.bf'] = 'Bitstream Vera Sans:bold'
plt.rc('text.latex', preamble=r'\usepackage{amsmath}')
mev_to_kj=96.485307/1000.
```

```python
phases=['Ih','II','III','V','VI']
DATA=dict()
```

```python
for phase in phases:
    DATA[phase]=dict()
    DATA[phase]=np.loadtxt('DATA/{}-pert-corr.dat'.format(phase))
```

```python
fig = plt.figure(figsize=(3.33,2.5))

for i, phase in enumerate(phases):
    fn='{}/{}-PIMD-correction.dat'.format(phase,phase)

   # axs[i].set_title(phase)
    
    data  = DATA[phase]
    mean=np.mean(data[:,2])#*96.4916/1000
    std=np.std(data[:,2])#*96.4916/1000
    plt.bar(i,mean,yerr= std,width=0.8,bottom=None, align='center',color=colors[i],edgecolor='k')

    plt.axhline(0, color='k', linewidth=0.5)
    plt.ylabel(r"$\Delta\Delta\mu^{\mathrm{MB-pol}}_{\mathrm{liq}\to \mathrm{ice}}$ (meV)")
plt.xticks(np.array(range(len(phases))), phases)#,size=6)
plt.tight_layout()
plt.savefig('figs/mean_nn-mbpol_correction.png',dpi=300)
```

```python
#print(corrs)
fig, ax = plt.subplots(2,3,figsize=(7.2,3.5))
axs=ax.flatten()
for i,key in enumerate(sorted(DATA.keys())):
    data=DATA[key]
    x=[]
    press=np.unique(data[:,1])
    for j,row in enumerate(data):
        
        index=np.where(np.abs(press-row[1])<0.001)[0][0]
        #print(index)
        axs[i].errorbar(index,row[2]*mev_to_kj,yerr=row[3]*mev_to_kj)
        axs[i].scatter(index,row[2]*mev_to_kj)#,yerr=row[3]+row[-1])
        axs[i].set_title(key)
        axs[i].set_ylim(bottom=-0.15,top=0.05)
        axs[-1].set_visible(False)
    axs[i].set_xticks(np.array(range(len(press))))
    axs[i].set_xticklabels([str(int(p)) for p in press])

for axi in ax[:,0]:
    axi.set_ylabel(r'$\Delta\Delta\mu_{\mathrm{liq}\to \mathrm{ice}}$ (kJ$\,$mol$^{-1}$)')
    
axs[-2].set_xlabel(r'P/atm')
axs[-3].set_xlabel(r'P/atm')
axs[-4].set_xlabel(r'P/atm')
plt.tight_layout()
plt.savefig('figs/CorrectionThermoPert.png',dpi=300, bbox_inches='tight',)
```

```python
fig, ax = plt.subplots(3,2,figsize=(7.2,7),sharey=True)
axs=ax.flatten()
for i,key in enumerate(phases):
    data=DATA[key].copy()

    data[:,2:]=data[:,2:]*mev_to_kj
    names=tuple(['{}atm {}K'.format(d[1],d[0]).replace('.0','').replace('000','k') for d in data])
    x=[]
    press=np.unique(data[:,1])
    axs[i].annotate('({}) {}'.format('abcdef'[i],key),xy=(0,1.05),xycoords='axes fraction')
    for j,row in enumerate(data):
        

        axs[i].bar(j, row[2], width=0.8, bottom=None, align='center',edgecolor='k',color=colors[j])
    #print(names[j])
        if j%2==0:
            axs[i].text(j,max([row[2],0])+0.1*mev_to_kj, names[j], rotation=70,size=6,rotation_mode='anchor',color=colors[j]) #color=colors[num[indices.index(N)]])
            #axs[i].annotate(xy=(x[j]-0.1, max([height[j],0])+0.1),text=names[j],rotation=45,size=5,color=colors[j])
        else:
            axs[i].text(j,min([row[2],0])-0.75*mev_to_kj, names[j], rotation=70,size=6,rotation_mode='anchor',color=colors[j])
    axs[i].set_ylim([-0.185,0.1])
    axs[i].set_xticks([])
    axs[i].axhline(0, color='k', linewidth=0.5)

    #axs[i].set_xticklabels([str(int(p)) for p in press])

for axi in ax[:,0]:
    axi.set_ylabel(r'$\Delta\mu^{\textrm{DNN@MB-pol}\to\textrm{MB-pol}}_{\textrm{liq}\to \textrm{ice}}$ (kJ$\,$mol$^{-1}$)')
fig.delaxes(axs[-1])
    

plt.tight_layout()
plt.savefig('figs/CorrectionThermoPert.png',dpi=300, bbox_inches='tight',)
plt.savefig('figs/CorrectionThermoPert.pdf',bbox_inches='tight',)
```
