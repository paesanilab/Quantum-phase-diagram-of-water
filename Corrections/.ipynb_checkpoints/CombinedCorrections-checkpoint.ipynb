# Combined correction of NQE and thermodynamic pertubation

```python
import numpy as np
import sys
import matplotlib.pylab as plt
from scipy import stats
from scipy.interpolate import interp1d
import scipy
from scipy import integrate
import scipy.integrate as integrate
from scipy.interpolate import UnivariateSpline
from scipy.optimize import curve_fit
from matplotlib import rc
import os
import matplotlib
from matplotlib.patches import Rectangle
SMALL_SIZE = 8
MEDIUM_SIZE = 10
BIGGER_SIZE = 12
plt.rc('font', size=MEDIUM_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
%config InlineBackend.figure_format = 'retina'
mev_to_kj=96.485307/1000.
colors=['#377eb8','#4daf4a','#e41a1c','#ff7f00','#a65628','#f781bf','#999999']*10 #' #4daf4a'
markers=["o"]*8+["s"]*8+['^']*8+['v']*8
!cp NQE/DATA/*PIMD* DATA/
!cp ThermoPert/DATA/*-pert-corr.dat DATA/
!rm DATA/*_*
```

```python
phases = ['Ih', 'II', 'III', 'V', 'VI']
```

```python
ncol=20
!rm DATA/*-sum-corr.dat 
for phase in phases:
    fn_sum='DATA/{}-sum-corr.dat'.format(phase)
    fn_nqe='DATA/{}-PIMD-correction.dat'.format(phase)
    fn_pert='DATA/{}-pert-corr.dat'.format(phase)
    data_nqe=np.loadtxt(fn_nqe)
    data_pert=np.loadtxt(fn_pert)
    P=data_pert[:,1]
    T=data_pert[:,0]
    with open(fn_sum,'w') as fp:
        fp.write('#temperature[K]'.ljust(ncol)+'pressure[atm]'.ljust(ncol)+'dmu[meV]'.ljust(ncol)+'dmu_error[meV]'.ljust(ncol)+'\n')#+'dmu_err[meV]'.ljust(ncol)
        index_pert=0
        for  (t,p) in zip(T,P):
            index_nqe=(np.where((data_nqe[:,0]==t)*(data_nqe[:,1]==p))[0][0])
            nqe=data_nqe[index_nqe]
            pert=data_pert[index_pert]
            fp.write(str(nqe[0]).ljust(ncol)+str(nqe[1]).ljust(ncol)+'{:.4f}'.format(nqe[2]+pert[2]).ljust(ncol)+'{:.4f}'.format((nqe[3]**2+pert[3]**2)**0.5).ljust(ncol)+'\n')
            index_pert+=1

            
            
        
#!cat DATA/*-sum-corr.dat        
        
```

```python
fig = plt.figure(figsize=(3.4,2.5))

width=0.2
offset=0.2
for phase in phases:
    for i, phase in enumerate(phases):
        fn_nqe='DATA/{}-PIMD-correction.dat'.format(phase)
        fn_pert='DATA/{}-pert-corr.dat'.format(phase)
        data_nqe=np.loadtxt(fn_nqe)*mev_to_kj
        data_pert=np.loadtxt(fn_pert)*mev_to_kj
        
        mean_nqe=data_nqe[:,2].mean()
        std_nqe=data_nqe[:,2].std()
        mean_pert=data_pert[:,2].mean()
        std_pert=data_pert[:,2].std()
        plt.axhline(0, color='k', linewidth=0.5)

        plt.bar(i-offset,mean_pert,yerr= std_pert*mev_to_kj,width=width,bottom=None, align='center',color=colors[0],edgecolor='k')
        plt.bar(i,mean_nqe,yerr= std_nqe,width=width,bottom=None, align='center',color=colors[1],edgecolor='k')
        plt.bar(i+offset,(mean_nqe+mean_pert),yerr= (std_nqe**2+std_pert**2)**0.5,width=width,bottom=None, align='center',color=colors[2],edgecolor='k')
#plt.ylim(top=1.25)
plt.xticks(np.array(range(len(phases))), phases)
plt.ylabel(r"$\Delta\mu_{\mathrm{liq}\to \mathrm{ice}}$ (kJ$\,$mol$^{-1}$)")
handles = [Rectangle((0,0),0.25,0.25, facecolor=c,edgecolor='k') for c in colors[:3]]   
legend=plt.legend(handles, [r'DNN@MB-pol$\to$MB-pol',r'Classical$\to$quantum','Net correction'],edgecolor="white",prop={'size': 7.},handletextpad=0.75,columnspacing=0.0,)#,frameon=False)
legend.get_frame().set_alpha(None)

plt.tight_layout()
for i in range(2):
    plt.axvline(x=i+0.5,color='k',alpha=0.3,lw=1)#,dashes=(4, 1))
    
ymin,ymax = plt.gca().get_ylim()
plt.ylim([ymin,ymax])
plt.vlines(x=2+0.5,color='k',alpha=0.3,lw=1,ymin=ymin,ymax=0.5*mev_to_kj)
plt.vlines(x=3+0.5,color='k',alpha=0.3,lw=1,ymin=ymin,ymax=0.5*mev_to_kj)

plt.savefig('figs/correction_sum.png',dpi=300, bbox_inches='tight', )
plt.savefig('figs/correction_sum.pdf', bbox_inches='tight', )
plt.savefig('figs/fig3.png',dpi=300, bbox_inches='tight', )
plt.savefig('figs/fig3.pdf', bbox_inches='tight', )
```
