# Correction from quantum nuclear effects 

```python
import numpy as np
import sys
import matplotlib.pylab as plt
from scipy import stats
from scipy.interpolate import interp1d
#import scipy.integrate as integrate
import scipy
from scipy import integrate
from scipy import integrate
mev_to_kj=96.485307/1000.

from scipy.interpolate import UnivariateSpline
from scipy.optimize import curve_fit
from matplotlib import rc
import os
import matplotlib
SMALL_SIZE = 8
MEDIUM_SIZE = 10
BIGGER_SIZE = 12
plt.rc('font', size=MEDIUM_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
%config InlineBackend.figure_format = 'retina'
plt.rcParams['text.usetex'] = True
matplotlib.rcParams['mathtext.fontset'] = 'custom'
matplotlib.rcParams['mathtext.rm'] = 'Bitstream Vera Sans'
matplotlib.rcParams['mathtext.it'] = 'Bitstream Vera Sans:italic'
matplotlib.rcParams['mathtext.bf'] = 'Bitstream Vera Sans:bold'
plt.rc('text.latex', preamble=r'\usepackage{amsmath}')
colors=['#377eb8','#4daf4a','#e41a1c','#984ea3','#ff7f00','#a65628','#f781bf','#999999']*10
markers=["o"]*8+["s"]*8+['^']*8+['v']*8
mev_to_kj=96.485307/1000.
```

```python
phases=['Ih','II','III','V','VI']
DATA=dict()
```

```python
for phase in phases:
    DATA[phase]=dict()
    data=np.loadtxt('{}/{}-kctv.dat'.format(phase,phase))
    data2 = np.loadtxt('{}/{}-kctv.dat'.format(phase,phase))
    P=data2[:,2]
    T=data2[:,1]
    PT=np.unique(np.vstack((P,T)),axis=1).T
    for P0, T0 in zip(P,T):
        data = np.loadtxt('{}/{}-kctv.dat'.format(phase,'Liq'))
        data2 = np.loadtxt('{}/{}-kctv.dat'.format(phase,phase))
        data=data[(P0==data[:,2])*(T0==data[:,1])]
        data2=data2[(P0==data2[:,2])*(T0==data2[:,1])]
        data=data[data[:,0].argsort()]
        data2=data2[data2[:,0].argsort()]
        
        # Integration variable                                                                                                                                                                                                                 
        x = data[:,0]
       
        # Integrand  
        try:
            y = 2*(data2[:,3]-data[:,3])/x
        except:
            print(phase,P0,T0)
           # print(data)
            #print(data2)
            continue
        # Error of integrand
        yerror=2*np.sqrt(data2[:,4]**2.+ data[:,4]**2.)
        
        # Add 0 correction to fully classical particles
        x=np.append(0,x)
        y=np.append(0,y)
        yerror=np.append(0.001,yerror)
        
        DATA[phase][str(T0)+'K '+str(P0)+'atm']=np.vstack((x,y,yerror)).T
  
```

## Integrand for quantum nuclear correction

```python
def func(x,b,c,d):
    return b*x+c*x*x+d*x**3 #+e*x**4
```

```python
fig, ax = plt.subplots(3,2,figsize=(7.2,7),sharey=True)
axs=ax.flatten()
for i, phase in enumerate(DATA.keys()):
    axs[i].annotate('({}) {}'.format('abcdef'[i],phase),xy=(0,1.05),xycoords='axes fraction')

    data=DATA[phase]
    for j, cond in enumerate(data.keys()):
        #axs[i].set_title(phase)
        axs[i].scatter(data[cond][:,0],data[cond][:,1]*mev_to_kj,color=colors[j],marker=markers[j],facecolor="none",label=cond.replace('000.0','k'))
        #xnew=np.linspace(0,1,10000)
    axs[i].legend(ncol=5,prop={'size': 4},handletextpad=0.25,labelspacing=0.7,columnspacing=0.5,frameon=False)
for axi in axs:
    axi.set_ylim(bottom=-0.5,top=1.1)
#  #   axi.legend()
for axi in ax[:,0]:
    axi.set_ylabel(r"$\frac{2\left(\left< T^{\mathrm{ice}}_{\mathrm{CV}} \right>-\left< T^{\mathrm{liq}}_{\mathrm{CV}} \right>\right)}{\sqrt{\tilde m/m}}$ (kJ$\,$mol$^{-1}$)")

for axi in ax[-1,:]:
    axi.set_xlabel(r'$\sqrt{m/\tilde m}$')
axs[3].set_xlabel(r'$\sqrt{m/\tilde m}$')

fig.delaxes(axs[-1])
plt.tight_layout()
plt.savefig('figs/integrand_nqe.png',dpi=300, bbox_inches='tight')
plt.savefig('figs/integrand_nqe.pdf', bbox_inches='tight')
```

## Integrate integrand to determine correction to chemical potential

```python
ncols=30
for i, phase in enumerate(DATA.keys()):
    fn='DATA/{}-PIMD-correction.dat'.format(phase,phase)
    with open(fn,'w') as fp:
        fp.write('#T[k]'.ljust(ncols) +'P[atm]'.ljust(ncols)+'dmu[ice-liq,simp,h2o]'.ljust(ncols)+'dmu_error[ice-liq,spline]'.ljust(ncols)+'dmu[ice-liq,simp,h2o]'.ljust(ncols)+'dmu_error[ice-liq,spline]'.ljust(ncols)+'\n')
        data=DATA[phase]
        for j, cond in enumerate((data.keys())):
            T=cond.split()[0].strip('K')
            P=cond.split()[1].strip('atm')
            
            x=data[cond][:,0]
            y=data[cond][:,1]
            yerror=data[cond][:,2]

            xnew=np.linspace(0,1,10000)
            xd2o=np.linspace(0,1/np.sqrt(2),10000)

            # Standard Simpson rule
            integral=integrate.simpson(y,x=x)
            
            error_simpson=0.5*(integrate.simpson(y+yerror/1.96,x=x)-integrate.simpson(y-yerror/1.96,x=x))
            #integral_d2o=integrate.simpson(y[:-1],x=x[:-1])

   
            # Numerical integral with spline
            spl = UnivariateSpline(data[cond][:,0], data[cond][:,1],w=1/yerror**2)
            spl1 = UnivariateSpline(data[cond][:,0], y-yerror/1.96,w=1/yerror**2)
            spl2 = UnivariateSpline(data[cond][:,0], y+yerror/1.96,w=1/yerror**2)
            
            integral2=np.trapz(spl(xnew), x=xnew)
            #integral2_d2o=np.trapz(spl(xd2o), x=xd2o)
            error_spline=0.5*(np.trapz(spl2(xnew), x=xnew)-np.trapz(spl1(xnew), x=xnew))
            #print(integral,integral2)
            fp.write(T.ljust(ncols)+P.ljust(ncols)+str(integral).ljust(ncols)+str(error_simpson).ljust(ncols)+str(integral2).ljust(ncols)+str(error_spline).ljust(ncols)+'\n')
!ls */*PIMD-correction.dat| xargs -l bash -c 'sort -gk2 -gk1 $0 --out $0'
```

```python
fig, ax = plt.subplots(3,2,figsize=(7.2,7))

axs=ax.flatten()
for i, phase in enumerate(phases):
    fn='{}/{}-PIMD-correction.dat'.format(phase,phase)

    axs[i].annotate('({}) {}'.format('abcdef'[i],phase),xy=(0,1.05),xycoords='axes fraction')
    
    data  = np.loadtxt(fn)
 #   print(data.shape)
    names=tuple(['{}atm {}K'.format(d[1],d[0]).replace('.0','').replace('000','k') for d in data])
    height = data[:,2].copy() 

    #height3 = data[:,4] 

    x = np.array(range(len(data)))
    #axs[i].set_xticks(np.array(range(len(data))), names)#,size=6,rotation=60
    axs[i].set_ylim(top=0.22,bottom=-0.2)
    
    for j in range(len(x)):
        axs[i].bar(x[j], height[j]*mev_to_kj, width=0.8, bottom=None, align='center',edgecolor='k',color=colors[j])
    #print(names[j])
        if j%2==0:
            axs[i].text(j,(max([height[j],0])+0.1)*mev_to_kj, names[j], rotation=70,size=6,rotation_mode='anchor',color=colors[j]) #color=colors[num[indices.index(N)]])
            #axs[i].annotate(xy=(x[j]-0.1, max([height[j],0])+0.1),text=names[j],rotation=45,size=5,color=colors[j])
        else:
            axs[i].text(j,(min([height[j],0])-1.)*mev_to_kj, names[j], rotation=70,size=6,rotation_mode='anchor',color=colors[j])

       #     axs[i].annotate(xy=(x[j]-0.5, min([height[j],0])-0.95),text=names[j],rotation=45,size=5,color=colors[j])
    axs[i].axhline(0, color='k', linewidth=0.5)
    axs[i].set_xticks([])
    
for axi in ax[:,0]:
    axi.set_ylabel(r"$\Delta\Delta\mu^{\mathrm{NQE}}_{\mathrm{liq}\to \mathrm{ice}}$ (kJ$\,$mol$^{-1}$)")
    

fig.delaxes(axs[-1])
plt.tight_layout()
plt.savefig('figs/nqe_correction.png',dpi=300,bbox_inches='tight')
plt.savefig('figs/nqe_correction.pdf',bbox_inches='tight')
```
