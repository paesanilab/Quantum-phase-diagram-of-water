# Distribution of order parameter

```python
import numpy as np
import matplotlib.pylab as plt
import matplotlib as mpl
#mpl.rcParams['figure.dpi'] = 300
%config InlineBackend.figure_format = 'retina'
ice='IceIh'
```

### Overlap between distributions

```python
data=np.loadtxt('results.txt')

plt.figure(figsize=(3.4,2))
plt.plot(data[:,0],data[:,1])
plt.xlabel(r'$\sigma$')
plt.ylabel('Overlap')
plt.tight_layout()
plt.show()
```

### Distributions

```python
hist_liquid = np.loadtxt('Liquid/histo')
hist_IceIh = np.loadtxt('IceIh/histo')

plt.figure(figsize=(3.4,2))
plt.plot(hist_IceIh[:,0],hist_IceIh[:,1],label='IceIh')
plt.plot(hist_liquid[:,0],hist_liquid[:,1],label='Liquid')
plt.xlim([0.3,1.1])
plt.annotate(r'$\sigma=0.07$',(0.7,4))

diff_val = np.abs(hist_liquid[:,1]-hist_IceIh[:,1])
diff_bin = hist_liquid[:,0]
diff_val = diff_val[(diff_bin>0.45) * (diff_bin<1.05)]
diff_bin = diff_bin[(diff_bin>0.45) * (diff_bin<1.05)]
min_order= diff_bin[diff_val.min()==diff_val]
#plt.axvline(x=min_order, color='k', linestyle='--')
plt.xlabel(r'$k_\chi$')
plt.ylabel(r'$P(k_\chi)$')
plt.legend(ncol=2)
plt.show()
```

```python
print('Liquid max:', hist_liquid[hist_liquid[:,1]==hist_liquid[:,1].max()][0][0])
```

```python
print('IceIh max:', hist_IceIh[hist_IceIh[:,1]==hist_IceIh[:,1].max()][0][0])
```

```python
print('Midpoint between distribitions:', min_order[0])
```
